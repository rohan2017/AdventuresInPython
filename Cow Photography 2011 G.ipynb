Brainstorm:
Go through an array and for every pair, find out their relative order by seeing what is their order in at least 3 pictures. 
Once you have the order of every pair you know the exact positions

```python
import os
import numpy as np
import operator

def readfile(file, basepath):
    base_path = basepath
    filename = file
    path_to_file = os.path.join(base_path, filename)
    testset = open(path_to_file , 'r')

    number = 0
    filelist = []
    for line in testset:
        line = line.strip()
        if line and line[0].isdigit():
            try:
                filelist.append(int(line))
            except ValueError:
                print("line not an integer")
    return filelist


def makematrix(filelist):
    cows = filelist[0]
    pictures = np.zeros((5, cows))
    count = 1
    for x in range(0, 5):
        for y in range(0, cows):
            pictures[x][y] = filelist[count]
            count += 1
    return pictures


def getpic(filelist):
    l = len(filelist)
    outputarray = []
    for x in range(1, int(((l-1)/5))+1):
        outputarray.append(filelist[x])
    return outputarray


def findpairs(inputarray):
    l = len(inputarray)
    pairnumbers = int(l/2 * (l - 1))
    print("The number of cows is:")
    print(l)
    print("The number of pairs is:")
    print(pairnumbers)
    pairs = np.zeros((pairnumbers, 2))
    
    counter = 0
    for x in range(0,l):
        for y in range(x+1,l):
            pairs[counter][0] = int(inputarray[x])
            pairs[counter][1] = int(inputarray[y])
            counter += 1
    return pairs


def findorders(pairmatrix, allpics):
    apcolumns = allpics.size/len(allpics)
    for x in range(0, len(pairmatrix)-1):
        num1 = 0
        num2 = 0
        num1 = pairmatrix[x][0]
        num2 = pairmatrix[x][1]
        num1p = 0
        num2p = 0
        for y in range(0, len(allpics)-1):
            for z in range(0, int(apcolumns)-1):
                if(allpics[y][z] == num1):
                    for a in range(0, int(apcolumns)-1):
                        if(allpics[y][a] == num2):
                            if(a > z):
                                num2p += 1
                            if(a < z):
                                num1p += 1
        if(num1p < num2p):
            print("pair order good")
        if(num2p < num1p):
            print("pair order bad, switching")
            placehold = pairmatrix[x][0]
            pairmatrix[x][0] = pairmatrix[x][1]
            pairmatrix[x][1] = placehold
        outputmatrix = pairmatrix 
    return outputmatrix # a 2 by n matrix with the pairs in order this time


def getfinalarray(orders, picture):
    ordercolumns = orders.size/len(orders)
    orderhash = {}
    for x in picture:
        orderhash[x] = 0
        for y in range(0, len(orders)):
            if(orders[y][0] == x):
                orderhash[x] = orderhash[x] - 1
            if(orders[y][1] == x):
                orderhash[x] = orderhash[x] + 1
    sortedlst = sorted(orderhash.items(), key=lambda kv: kv[1])
    return sortedlst


dataarray = readfile("I.4", "C:/Users/rohan/OneDrive/Documents/USACODatasets/photo")
print("File contents:")
print(dataarray)
firstpic = getpic(dataarray)
print("A picture looks like:")
print(firstpic)
pairs = findpairs(firstpic)
print("Here are the pairs:")
print(pairs)
matrix = makematrix(dataarray)
print("The data matrix is:")
print(matrix)
orderedpair = findorders(pairs, matrix)
print("Now here are the pairs correctly ordered")
print(orderedpair)
sortedarr = getfinalarray(orderedpair, firstpic)
print("Now here is the original order")
print(sortedarr)
```
